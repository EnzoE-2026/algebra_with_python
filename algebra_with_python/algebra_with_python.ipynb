# Proportions

## $\frac{n1}{d1}$ = $\frac{n2}{d2}$

```python
def solve_proportion():
    """
    Solve a proportion with one unknown value (0).
    Expects 4 integers: n1, d1, n2, d2
    Exactly one value should be zero, representing the unknown.

    Returns:
        float: the calculated unknown value
        None: if the input is proportionate (no unknown)
        -1: if the proportion is invalid
    """

    # Input loop
    while True:
        try:
            nums = []
            print("Enter 4 integers for n1, d1, n2, d2 (use 0 for the unknown):")
            for _ in range(4):
                nums.append(int(input("Enter a valid integer: ")))

            # Validation
            if len(nums) != 4:
                print("You must enter exactly 4 numbers.\n")
                continue

            zero_count = nums.count(0)
            if zero_count != 1:
                print("Exactly one value should be 0 to represent the unknown.\n")
                continue #starts loop from beginning

            n1, d1, n2, d2 = nums
            break #breaks out of loop

        except ValueError:
            print("All inputs must be integers.\n")

    # Solve for the unknown
    try:
        if n1 == 0:
            answer = (d1 * n2) / d2
        elif d1 == 0:
            answer = (n1 * d2) / n2
        elif n2 == 0:
            answer = (n1 * d2) / d1
        elif d2 == 0:
            answer = (d1 * n2) / n1
        else:
            # No unknown; check if proportion holds
            if n1 * d2 == d1 * n2:
                print("Proportionate ✅")
                return None
            else:
                print("Non-proportionate 🚫")
                return -1

        print(f"The unknown value is: {answer}")
        return answer

    except ZeroDivisionError:
        print("Error: division by zero occurred. Check your inputs.")
        return None
```
